```python
import requests
import json
import subprocess
import csv
import pandas as pd
```

```python
fields = [
    "file_name",
    "file_id",
    "case_id",
    "cases.submitter_id",
    "cases.samples.sample_type",
    "cases.disease_type",
    "cases.project.project_id"
    ]

fields = ",".join(fields)

files_endpt = "https://api.gdc.cancer.gov/files"

# This set of filters is nested under an 'and' operator.
filters = {
    "op": "and",
    "content":[
        {
        "op": "in",
        "content":{
            "field": "files.data_format",
            "value": ["svs"]
            }
        },
    ]
}

# A POST is used, so the filter parameters can be passed directly as a Dict object.
params = {
    "filters": filters,
    "fields": fields,
    "format": "CSV",
    "size": "30075"
    }
```

```python
# The parameters are passed to 'json' rather than 'params' in this case
response = requests.post(files_endpt, headers = {"Content-Type": "application/json"}, json = params)
data = response.content.decode("utf-8")
```

```python
# open the file in the write mode
with open('slides.csv', 'w', newline='') as csvfile:
    
    # Create a CSV writer object
    writer = csv.writer(csvfile)

    # Split the data string into rows
    rows = data.strip().split('\n')

    # Write each row to the CSV file
    for row in rows:
        if("\"" not in row[0]):
            r = row.split(',')
            writer.writerow(r)
        else:
            r = row.split(',')
            restR =r[2:]
            firstR = [r[0] + r[1]]
            firstR[0] = firstR[0].strip("\"")
            r = firstR + restR
            writer.writerow(r)
```

```python
slidesdf = pd.read_csv(r'slides.csv')
slidesdf.drop_duplicates(subset=['cases.0.submitter_id'], inplace=True)

slidesdf.to_csv('slides.csv', index=False)
slidesdf
```

```python
fields = [
    "file_name",
    "file_id",
    "cases.submitter_id",
    ]

fields = ",".join(fields)
# This set of filters is nested under an 'and' operator.
filters = {
    "op": "and",
    "content":[
        {
        "op": "in",
        "content":{
            "field": "files.data_format",
            "value": ["PDF"]
            }
        },
    ]
}


# A POST is used, so the filter parameters can be passed directly as a Dict object.
params = {
    "filters": filters,
    "fields": fields,
    "format": "CSV",
    "size": "11211"
    }
```

```python
# open the file in the write mode
with open('pdfs.csv', 'w', newline='') as csvfile:
    
     # Create a CSV writer object
    writer = csv.writer(csvfile)

    # Split the data string into rows
    rows = data.strip().split('\n')

    # Write each row to the CSV file
    for row in rows:
        if("\"" not in row[0]):
            r = row.split(',')
            writer.writerow(r)
        else:
            r = row.split(',')
            restR =r[2:]
            firstR = [r[0] + r[1]]
            firstR[0] = firstR[0].strip("\"")
            r = firstR + restR
            writer.writerow(r)

```

```python
pdfsdf = pd.read_csv(r'pdfs.csv')
pdfsdf.drop_duplicates(subset=['cases.0.submitter_id'], inplace=True)
pdfsdf.to_csv('pdfs.csv', index=False)
pdfsdf
```

```python
combinedf = pd.merge(slidesdf, pdfsdf, on='cases.0.submitter_id')
combinedf
```

Downloading Manifest 

```python
command = "curl --remote-name --remote-header-name 'https://api.gdc.cancer.gov/v0/manifest/"
uuids = []
for i in range(len(combinedf))):
    svsId = combinedf.loc[i, 'file_id_x']
    pdfId = combinedf.loc[i, 'file_id_y']
    uuids.append(svsId)
    uuids.append(pdfId)
uuids = ','.join(uuids)
command = command + uuids + "'"
command

output = subprocess.check_output(command, shell=True)

# Print the output of the command
print(output.decode())
    
```
